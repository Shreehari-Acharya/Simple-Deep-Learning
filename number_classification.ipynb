```python
import tensorflow
import keras
from keras import Sequential
from keras.layers import Dense, Flatten, Input
import numpy as np
from PIL import Image
```

```python
(x_train,y_train),(x_test,y_test) = keras.datasets.mnist.load_data()

# bring all the values between 0 and 1. this will help compute weights faster
x_train = x_train/255
x_test = x_test/255
```

```python
model = Sequential([
    Input(shape=(28, 28)),
    Flatten(),
    Dense(128, activation="relu"),
    Dense(64, activation="relu"),
    Dense(32, activation="relu"),
    Dense(10, activation="softmax")
])

model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')
model.fit(x_train,y_train,epochs=20,validation_split=0.2)
```

```python
from ipycanvas import Canvas
from IPython.display import display
from ipywidgets import Button, VBox, Layout
```

```python
from google.colab import output
output.enable_custom_widget_manager()
```

```python
def canvas_to_28x28(canvas):
    img = Image.fromarray(canvas.get_image_data())
    img = img.convert("L")
    img = img.resize((28, 28))

    arr = np.array(img) / 255.0
    return arr
```

```python
def predict_digit():
  img_array = canvas_to_28x28(canvas)
  prediction = model.predict(img_array.reshape(1,28,28)).argmax(axis=1)
  print(f"The prediction from the model is {prediction[0]}")
  return
```

```python
canvas = Canvas(
    width=280,
    height=280,
    sync_image_data=True
)
canvas.fill_style = "black"
canvas.fill_rect(0, 0, 280, 280)

canvas.stroke_style = "white"
canvas.line_width = 20

drawing = False

def on_mouse_down(x, y):
    global drawing
    drawing = True
    canvas.begin_path()
    canvas.move_to(x, y)

def on_mouse_move(x, y):
    if not drawing:
        return
    canvas.line_to(x, y)
    canvas.stroke()

def on_mouse_up(x, y):
    global drawing
    drawing = False

canvas.on_mouse_down(on_mouse_down)
canvas.on_mouse_move(on_mouse_move)
canvas.on_mouse_up(on_mouse_up)

# --- Button ---
predict_btn = Button(
    description="Predict",
    button_style="success"
)
clear_btn = Button(
    description="Clear",
    button_style="danger"
)

def on_clear_clicked(b):
    canvas.clear()
    canvas.fill_style = "black"
    canvas.fill_rect(0, 0, 280, 280)

def on_predict_clicked(b):
    predict_digit()

predict_btn.on_click(on_predict_clicked)
clear_btn.on_click(on_clear_clicked)

# --- Layout ---
display(VBox([canvas, predict_btn, clear_btn], layout=Layout(width="320px")))
```
